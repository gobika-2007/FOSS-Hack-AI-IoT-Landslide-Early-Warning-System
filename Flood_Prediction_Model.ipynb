```python
data = pd.read_csv("flood_dataset.csv")
```

```python
print(data.head())
```

```python
# Convert Landslide values to binary (Risk / No Risk)
data['Risk'] = data['Landslides'].apply(lambda x: 1 if x >= 5 else 0)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

# Input features
X = data.drop(['Landslides', 'Risk'], axis=1)

# Target
y = data['Risk']

# Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

# Train model
model = RandomForestClassifier()
model.fit(X_train, y_train)

print("Model trained successfully!")
```

```python
import pandas as pd

# Take sample
sample = X_test.iloc[0]

# Convert properly with index (IMPORTANT)
sample_df = pd.DataFrame(sample).T

# Ensure column names match
sample_df.columns = X.columns

# Predict
prediction = model.predict(sample_df)

print("Prediction:", prediction)

if prediction[0] >= 5:
    print("⚠️ High Landslide Risk!")
else:
    print("✅ Low Risk / Safe")
```

```python
from sklearn.metrics import accuracy_score

y_pred = model.predict(X_test)
print("New Accuracy:", accuracy_score(y_test, y_pred))
```
